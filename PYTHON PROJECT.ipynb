```python
import matplotlib.pyplot as plt
age_col = 'Age'
df[age_col].value_counts().plot(kind='bar',color='orange')
plt.title("Age Group Distribution")
plt.show()
```

```python
vote_col = 'Your Vote'

df[vote_col].value_counts().plot(
    kind='pie',
    labels=df[vote_col].value_counts().index,
    labeldistance=0.5,
    autopct=None,
    ylabel=''
)

plt.title("Your Vote")
plt.show()
```

```python
vote_col = 'Your Prediction'

df[vote_col].value_counts().plot(
    kind='pie',
    labels=df[vote_col].value_counts().index,
    labeldistance=0.5,
    autopct=None,
    ylabel=''
)

plt.title("Prediction")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#df = pd.read_csv("2026 - ELECTION.csv")

age_col = 'Age'
vote_col = 'Your Vote'

age_vote = pd.crosstab(df[age_col], df[vote_col])

age_vote.plot(kind='bar')
plt.title("Age Group vs Voting Preference (2026)")
plt.xlabel("Age Group")
plt.ylabel("Number of Responses")
plt.show()
```

```python
#import pandas as pd
#import matplotlib.pyplot as plt

#df = pd.read_csv("2026 - ELECTION.csv")

gender_col = 'Gender'
vote_col = 'Your Vote'

gender_vote = pd.crosstab(df[gender_col], df[vote_col])

gender_vote.plot(kind='bar')
plt.title("Gender vs Voting Preference")
plt.xlabel("Gender")
plt.ylabel("Voting Preference")
plt.xticks(rotation=0)
plt.show()
```

```python


gender_col = 'Gender'
vote_col = 'Your Prediction'

gender_vote = pd.crosstab(df[gender_col], df[vote_col])

gender_vote.plot(kind='bar')
plt.title("Gender vs Voting Predection")
plt.xlabel("Gender")
plt.ylabel("Voting Predection")
plt.xticks(rotation=0)
plt.show()
```
